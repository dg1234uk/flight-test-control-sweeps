# Control Sweeps

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
stick_deflection = np.array([0, 25, 50, 100])  # Stick deflection (%)
aileron_deflection = np.array(
    [0, 10, 20, 40]
)  # Aileron deflection (degrees, scaled to ±30 max)
elevator_deflection = np.array(
    [0, 9, 18, 36]
)  # Elevator deflection (degrees, scaled to ±40 max)

# Maximum stick deflection in cm
max_deflection_cm_lat = 7  # Left/Right axis
max_deflection_cm_lon = 12  # Fore/Aft axis


# Conversion functions for left/right (aileron) axis
def percent_to_cm_lat(x):
    return (x / 100) * max_deflection_cm_lat


def cm_to_percent_lat(x):
    return (x / max_deflection_cm_lat) * 100


# Conversion functions for fore/aft (elevator) axis
def percent_to_cm_lon(x):
    return (x / 100) * max_deflection_cm_lon


def cm_to_percent_lon(x):
    return (x / max_deflection_cm_lon) * 100


# Function to plot aileron deflection
def plot_aileron_deflection(save_path=None):
    fig, ax = plt.subplots(figsize=(6, 6), dpi=300)
    ax.plot(stick_deflection, aileron_deflection, "o-", label="Right Deflection")
    ax.plot(-stick_deflection, -aileron_deflection, "s-", label="Left Deflection")
    ax.axhline(0, color="gray", linestyle="--", linewidth=0.5)
    ax.axvline(0, color="gray", linestyle="--", linewidth=0.5)
    ax.set_title("Stick Deflection vs Aileron Deflection")
    ax.set_xlabel("Stick Deflection [%]")
    ax.set_ylabel("Aileron Deflection [degrees]")
    ax.legend()
    ax.grid(True)

    # Add second x-axis for stick deflection in cm (Left/Right)
    ax_secondary = ax.secondary_xaxis(
        "top", functions=(percent_to_cm_lat, cm_to_percent_lat)
    )
    ax_secondary.set_xlabel("Stick Deflection [cm]")

    if save_path:
        plt.savefig(save_path, bbox_inches="tight")
    plt.close(fig)


# Function to plot elevator deflection
def plot_elevator_deflection(save_path=None):
    fig, ax = plt.subplots(figsize=(6, 6), dpi=300)
    ax.plot(stick_deflection, elevator_deflection, "o-", label="Forward Deflection")
    ax.plot(-stick_deflection, -elevator_deflection, "s-", label="Aft Deflection")
    ax.axhline(0, color="gray", linestyle="--", linewidth=0.5)
    ax.axvline(0, color="gray", linestyle="--", linewidth=0.5)
    ax.set_title("Stick Deflection vs Elevator Deflection")
    ax.set_xlabel("Stick Deflection [%]")
    ax.set_ylabel("Elevator Deflection [degrees]")
    ax.legend()
    ax.grid(True)

    # Add second x-axis for stick deflection in cm (Fore/Aft)
    ax_secondary = ax.secondary_xaxis(
        "top", functions=(percent_to_cm_lon, cm_to_percent_lon)
    )
    ax_secondary.set_xlabel("Stick Deflection [cm]")

    if save_path:
        plt.savefig(save_path, bbox_inches="tight")
    plt.close(fig)


# Create the combined subplot figure
plt.figure(figsize=(12, 6), dpi=300)

# Aileron deflection subplot
ax1 = plt.subplot(1, 2, 1)
ax1.plot(stick_deflection, aileron_deflection, "o-", label="Right Deflection")
ax1.plot(-stick_deflection, -aileron_deflection, "s-", label="Left Deflection")
ax1.axhline(0, color="gray", linestyle="--", linewidth=0.5)
ax1.axvline(0, color="gray", linestyle="--", linewidth=0.5)
ax1.set_title("Stick Deflection vs Aileron Deflection")
ax1.set_xlabel("Stick Deflection [%]")
ax1.set_ylabel("Aileron Deflection [degrees]")
ax1.legend()
ax1.grid(True)

# Add second x-axis for stick deflection in cm (Left/Right)
ax1_secondary = ax1.secondary_xaxis(
    "top", functions=(percent_to_cm_lat, cm_to_percent_lat)
)
ax1_secondary.set_xlabel("Stick Deflection [cm]")

# Elevator deflection subplot
ax2 = plt.subplot(1, 2, 2)
ax2.plot(stick_deflection, elevator_deflection, "o-", label="Forward Deflection")
ax2.plot(-stick_deflection, -elevator_deflection, "s-", label="Aft Deflection")
ax2.axhline(0, color="gray", linestyle="--", linewidth=0.5)
ax2.axvline(0, color="gray", linestyle="--", linewidth=0.5)
ax2.set_title("Stick Deflection vs Elevator Deflection")
ax2.set_xlabel("Stick Deflection [%]")
ax2.set_ylabel("Elevator Deflection [degrees]")
ax2.legend()
ax2.grid(True)

# Add second x-axis for stick deflection in cm (Fore/Aft)
ax2_secondary = ax2.secondary_xaxis(
    "top", functions=(percent_to_cm_lon, cm_to_percent_lon)
)
ax2_secondary.set_xlabel("Stick Deflection [cm]")

# Adjust layout and save combined plot
plt.tight_layout()
plt.savefig(
    "./output/plots/control_sweeps_combined_deflection_plot.png", bbox_inches="tight"
)
plt.show()

# Save individual plots
plot_aileron_deflection("./output/plots/control_sweeps_aileron_deflection_plot.png")
plot_elevator_deflection("./output/plots/control_sweeps_elevator_deflection_plot.png")
```
